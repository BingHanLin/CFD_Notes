### Function

```julia
f(x::Float64) = x^2/sin(x)

df(x::Float64) = x*(2-x*cot(x))*csc(x)

error(exactSol::Float64, estimateSol::Float64) = abs(exactSol-estimateSol)
```

### Numerical Scheme

```julia
centerDiff(x::Float64, h::Float64) = (f(x+h)-f(x-h))/(2*h)
```

### Prepare Data

```julia
errorData = Array{Float64,1}(undef,0)

hData = exp10.(-15:1:-1)
for h in hData
    xValue = 1.0
    exactSol = df(xValue)
    estimateSol = centerDiff(xValue,h)
    oneError = error(exactSol,estimateSol)
    push!(errorData,oneError)
end
```

```julia
using Plots

plot(hData, errorData,  xlabel="h size", ylabel="error",markershape = :xcross,  markersize = 3, markercolor = :black, linestyle = :dashdot,  legend = false,  xaxis=:log, yaxis=:log)
```
